```python
# Importing all the required libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
pd.set_option('display.max_colwidth',2000)
pd.set_option('display.max_columns',200)
```

```python
# Reading the data from excel file and creating a dataframe

df = pd.read_excel('Fyttlyf_Website_Data.xlsx')
df
```

```python
# Renaming all the Columns for convinience

df.columns=['evnt_dt','event','geo_cntry','sessn_id','visitor_id','payload','page_name','devc_name','browser_type','traffic_source','Level_1','Level_2','Level_3','Level_4','Level_5']
df
```

```python
# Creating 'temp' as a copy of our dataframe

temp = df.copy()
temp
```

# Data Cleaning

```python
# Drop all the rows with null values

temp.dropna(axis='rows',inplace=True)
```

```python
# Dropping all the duplicate visistor_id rows 

temp.drop_duplicates('visitor_id',inplace=True)
temp['traffic_source'].value_counts()
```

```python
# a. To get the output that looks like the below image. 
# b. x1 in the image could be read as Unique count of visitors of Level 2 / Unique count of visitors of Level 1. 
# c. The function should take the rows (in this case Traffic source) as an argument i.e., we should be able to provide “devc_name” or “browser_type” as an argument to get the same table
```

```python
# Function for 'traffic_source'

table = temp.groupby('traffic_source', as_index=False)['Level_1','Level_2','Level_3','Level_4','Level_5'].sum()
table['Level_2/Level_1'] = table['Level_2']/table['Level_1']
table['Level_3/Level_2'] = table['Level_3']/table['Level_2']
table['Level_4/Level_3'] = table['Level_4']/table['Level_3']
table['Level_5/Level_4'] = table['Level_5']/table['Level_4']
table.drop(['Level_1','Level_2','Level_3','Level_4','Level_5'],axis=1,inplace=True)
table
```

```python
# Function for 'devc_name'

table = temp.groupby('devc_name', as_index=False)['Level_1','Level_2','Level_3','Level_4','Level_5'].sum()
table['Level_2/Level_1'] = table['Level_2']/table['Level_1']
table['Level_3/Level_2'] = table['Level_3']/table['Level_2']
table['Level_4/Level_3'] = table['Level_4']/table['Level_3']
table['Level_5/Level_4'] = table['Level_5']/table['Level_4']
table.drop(['Level_1','Level_2','Level_3','Level_4','Level_5'],axis=1,inplace=True)
table
```

```python
# Function for 'browser_type'

table = temp.groupby('browser_type', as_index=False)['Level_1','Level_2','Level_3','Level_4','Level_5'].sum()
table['Level_2/Level_1'] = table['Level_2']/table['Level_1']
table['Level_3/Level_2'] = table['Level_3']/table['Level_2']
table['Level_4/Level_3'] = table['Level_4']/table['Level_3']
table['Level_5/Level_4'] = table['Level_5']/table['Level_4']
table.drop(['Level_1','Level_2','Level_3','Level_4','Level_5'],axis=1,inplace=True)
table
```

```python
# Creating a second temperory copy of our dataframe & Dropping all the duplicate session id

temp2 = df.copy()

temp2.drop_duplicates('sessn_id',inplace=True)
temp2
```

```python
# Crearing dummy variables to count no. of clicks and impressions

dummies = pd.get_dummies(temp2.event, prefix='event')
dummies
```

```python
# Adding the dummy columns in our 'temp2' dataframe

temp2 = pd.concat([temp2,dummies],axis='columns')
temp2
```

```python
#  a. What is the CTR (Click-through rate) at geo_cntry level. (CTR = count distinct sessn_id where event = click / count distinct sessn_id where event = impressions )

# CTR = Click through rate
# There are only two countries that is providing the CTR

CTR = temp2.groupby('geo_cntry', as_index=False)['event_Click','event_Impression'].sum()
CTR['CTR'] = CTR['event_Click']/CTR['event_Impression']
CTR
```
